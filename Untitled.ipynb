```python
# Import necessary libraries
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define a custom date parser function for the specific date format
date_parser = lambda x: pd.to_datetime(x, format="%d-%m-%y %H:%M")

# Load the datasets with the specified date format
actual_consumption = pd.read_csv(
    "Actual consumption.csv", 
    parse_dates=["Start date"], 
    date_format="%d-%m-%y %H:%M"
)
forecasted_consumption = pd.read_csv(
    "Forecasted consumption.csv", 
    parse_dates=["Start date"], 
    date_format="%d-%m-%y %H:%M"
)

# Set 'Start date' as the index and remove 'End date' column
actual_consumption.set_index("Start date", inplace=True)
forecasted_consumption.set_index("Start date", inplace=True)
actual_consumption.drop(columns=["End date"], inplace=True)
forecasted_consumption.drop(columns=["End date"], inplace=True)
```

```python
# Display the first few rows of each dataframe to verify the changes
print("Actual Consumption Data:")
actual_consumption.head()
```

```python
print("\nForecasted Consumption Data:")
forecasted_consumption.head()
```

```python
# Merge the datasets on the index to ensure only matching dates are considered
combined_data = actual_consumption.join(forecasted_consumption, lsuffix='_Actual', rsuffix='_Forecast', how="inner")

# Display the first few rows of the combined data to verify
print("Combined Data:")
combined_data.head()
```

```python
combined_data.dtypes
```

```python
# Remove commas and convert columns to float type
combined_data["Total (grid load) [MWh]_Actual"] = combined_data["Total (grid load) [MWh]_Actual"].str.replace(",", "").astype(float)
combined_data["Residual load [MWh]_Actual"] = combined_data["Residual load [MWh]_Actual"].str.replace(",", "").astype(float)
combined_data["Hydro pumped storage [MWh]"] = combined_data["Hydro pumped storage [MWh]"].str.replace(",", "").astype(float)
combined_data["Total (grid load) [MWh]_Forecast"] = combined_data["Total (grid load) [MWh]_Forecast"].str.replace(",", "").astype(float)
combined_data["Residual load [MWh]_Forecast"] = combined_data["Residual load [MWh]_Forecast"].str.replace(",", "").astype(float)

# Verify the data types
print(combined_data.dtypes)
```

```python
# Step 1: Basic Summary Statistics
summary_stats = combined_data.describe()
print("\nSummary Statistics:")
summary_stats
```

```python
# Step 2: Data Distributions
# Histograms for Actual vs Forecasted Total (grid load) and Residual load
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(combined_data["Total (grid load) [MWh]_Actual"], color="blue", label="Actual", kde=True)
sns.histplot(combined_data["Total (grid load) [MWh]_Forecast"], color="orange", label="Forecast", kde=True)
plt.legend()
plt.title("Distribution of Total (grid load) [MWh]")

plt.subplot(1, 2, 2)
sns.histplot(combined_data["Residual load [MWh]_Actual"], color="blue", label="Actual", kde=True)
sns.histplot(combined_data["Residual load [MWh]_Forecast"], color="orange", label="Forecast", kde=True)
plt.legend()
plt.title("Distribution of Residual load [MWh]")

plt.tight_layout()
plt.show()
```

```python
# Step 3: Calculate Differences
combined_data["Total_diff"] = combined_data["Total (grid load) [MWh]_Actual"] - combined_data["Total (grid load) [MWh]_Forecast"]
combined_data["Residual_diff"] = combined_data["Residual load [MWh]_Actual"] - combined_data["Residual load [MWh]_Forecast"]

# Percentage Differences
combined_data["Total_diff_pct"] = (combined_data["Total_diff"] / combined_data["Total (grid load) [MWh]_Actual"]) * 100
combined_data["Residual_diff_pct"] = (combined_data["Residual_diff"] / combined_data["Residual load [MWh]_Actual"]) * 100
```

```python
# Step 4: Plot Differences Over Time
plt.figure(figsize=(14, 8))

plt.subplot(2, 1, 1)
plt.plot(combined_data.index, combined_data["Total_diff"], label="Total (grid load) Difference [MWh]", color="purple")
plt.title("Difference in Total (grid load) Over Time")
plt.ylabel("Difference in MWh")
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(combined_data.index, combined_data["Residual_diff"], label="Residual load Difference [MWh]", color="brown")
plt.title("Difference in Residual load Over Time")
plt.ylabel("Difference in MWh")
plt.legend()

plt.tight_layout()
plt.show()

```

```python
# Step 5: Summary of Differences
diff_summary_stats = combined_data[["Total_diff", "Residual_diff", "Total_diff_pct", "Residual_diff_pct"]].describe()
print("\nSummary Statistics of Differences:")
print(diff_summary_stats)
```

```python
# Plot for Actual Consumption
plt.figure(figsize=(16, 8))
plt.plot(actual_consumption.index, actual_consumption["Total (grid load) [MWh]"], label="Total (grid load) [MWh]", color="blue")
plt.plot(actual_consumption.index, actual_consumption["Residual load [MWh]"], label="Residual load [MWh]", color="green")
plt.xlabel("Date")
plt.ylabel("Consumption (MWh)")
plt.title("Actual Consumption Over Time")
plt.legend()
plt.grid(True)
plt.show()
```
